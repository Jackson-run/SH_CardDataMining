```python
def cal_score(pre,y):
    cost = np.sqrt(np.power((pre-y),2))
    return cost
```

```python
cost = cal_score(pre,y_test)
print np.mean(cost)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(rnd_clf,pre.reshape(-1,1),y_test.reshape(-1,1),cv=5,scoring="neg_mean_squared_error")
rmse_scores = np.sqrt(-scores)
```
